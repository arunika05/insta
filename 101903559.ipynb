```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
train_df = pd.read_csv('/kaggle/input/how-to-be-instagram-famous-with-data-science/dev.csv')
train_df.head()
```

```python
train_df.shape
```

```python
train_df.info()
```

```python
train_df['has_new_comments'].unique()
```

```python
x_data = train_df.iloc[:,:-2]
y_data = train_df.iloc[:,-2]
```

```python
x_data = pd.get_dummies(x_data, columns=['day_posted', 'current_day'], dtype='int64')
```

```python
y_data.shape
```

```python
for i in range(len(y_data)):
    if y_data[i] == 'yes':
        y_data[i] = 1
    else:
        y_data[i] = 0
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

```python
y_train = y_train.astype('int64')
```

```python
y_val = y_val.values.reshape(1,-1)
y_val = y_val.flatten()
```

```python
def get_accuracy(pred, real):
    count=0
    for i in range(len(pred)):
        if pred[i] == real[i]:
            count = count+1
    return count/len(pred)
```

```python
# print(get_accuracy(predict, y_val))
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)
```

```python
pred_rf = rf.predict(x_val)
```

```python
print(get_accuracy(pred_rf, y_val))
```
